```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['Label_Sex'] = le.fit_transform(df.Sex)
```

```python
df['ChestPainType'] = le.fit_transform(df.ChestPainType)
df['RestingECG'] = le.fit_transform(df.RestingECG)
df['ExerciseAngina'] = le.fit_transform(df.ExerciseAngina)
df['ST_Slope'] = le.fit_transform(df.ST_Slope)
```

```python
df
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
df.drop(['Label_Sex'],axis=1,inplace=True)
X = df.drop(['HeartDisease'],axis=1)
y = df.HeartDisease
```

```python
X
```

```python
y
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y,test_size=0.3)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
from sklearn.ensemble import BaggingClassifier
bagged_svc = BaggingClassifier(base_estimator=SVC(),n_estimators=100,oob_score=True)
```

```python
bagged_svc.fit(X_train,y_train)
bagged_svc.oob_score_
```

```python
bagged_svc.score(X_test,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
model2 = DecisionTreeClassifier()
model2.fit(X_train,y_train)
model2.score(X_test,y_test)
```

```python
bagged_DecTree = BaggingClassifier(DecisionTreeClassifier(),n_estimators=100,oob_score=True)
```

```python
bagged_DecTree.fit(X_train,y_train)
bagged_DecTree.score(X_test,y_test)
```

```python
bagged_DecTree.oob_score_
```
