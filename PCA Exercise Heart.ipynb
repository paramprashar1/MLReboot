```python
from sklearn.preprocessing import StandardScaler
stdScaler = StandardScaler()
```

```python
dfFinal = stdScaler.fit_transform(df[['RestingBP','Cholesterol','MaxHR','Oldpeak']])
```

```python
dfFinal = pd.DataFrame(dfFinal,columns=['Scaled_RestingBP','Scaled_Cholesterol','Scaled_MaxHR','Scaled_Oldpeak'])
```

```python
dfFinal = pd.concat([df, dfFinal], axis=1)
```

```python
dfFinal
```

```python
dfFinal.drop(['RestingBP','Cholesterol','MaxHR','Oldpeak'],axis=1,inplace=True)
```

```python
dfFinal
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
model_params={
    'logistic_regression':{
        'model' : LogisticRegression(solver='liblinear',multi_class='auto'),
        'params' : {
            'C':[1,5,10,19,150]
        }
    },
    'random_forest':{
        'model' :RandomForestClassifier(),
        'params' : {
            'n_estimators':[1,5,10,25,56]
        }
    },
    'svm':{
        'model' : SVC(gamma='auto'),
        'params' : {
            'C':[1,5,10,50,60,100],
            'kernel':['rbf','linear']
        }
    }
    
}
```

```python
dfFinal
```

```python
target = dfFinal.HeartDisease
target
```

```python
dfFinal.drop(['HeartDisease'],axis=1,inplace=True)
dfFinal
```

```python
scores =[]
for model_name,mp in model_params.items():
    clf = GridSearchCV(mp['model'],mp['params'],cv=5,return_train_score=False)
    clf.fit(dfFinal,target)
    scores.append({
        'model':model_name,
        'best_score':clf.best_score_,
        'best_params':clf.best_params_
    })
    
df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.95)
X=dfFinal
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
scores =[]
for model_name,mp in model_params.items():
    clf = GridSearchCV(mp['model'],mp['params'],cv=5,return_train_score=False)
    clf.fit(X_pca,target)
    scores.append({
        'model':model_name,
        'best_score':clf.best_score_,
        'best_params':clf.best_params_
    })
    
df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```
