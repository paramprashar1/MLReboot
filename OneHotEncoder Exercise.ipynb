```python
import pandas as pd
df = pd.read_csv("C:/Users/HP/Desktop/carprices.csv")
```

```python
df
```

```python
dummies = pd.get_dummies(df[['Car Model']])
dummies
```

```python
final = pd.concat([df,dummies],axis='columns')
final
```

```python
final = final.drop(['Car Model','Car Model_Mercedez Benz C class'],axis='columns')
final
```

```python
X = final.drop(['Sell Price($)'],axis='columns')
X
```

```python
y = final[['Sell Price($)']]
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(X,y)
```

```python
reg.score(X,y)
```
